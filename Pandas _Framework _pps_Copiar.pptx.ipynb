```python
# Divide a coluna 'Genres' em várias colunas e empilha-as em uma única Série
stack_genres = df_apps['Genres'].str.split(';', expand=True).stack()

print(f'Temos agora uma única coluna (Série) com a forma: {stack_genres.shape}')
```

```python
num_genres = stack_genres.value_counts()
print(f'Número de gêneros únicos: {len(num_genres)}')
print('Top 10 Gêneros:')
print(num_genres.head(10))
```

```python
import plotly.express as px
```

```python
bar = px.bar(x = num_genres.index[:15], # nomes dos gêneros (índice da série)
             y = num_genres.values[:15], # contagem dos gêneros (valores da série)
             title='Top 15 Gêneros na Google Play Store',
             hover_name=num_genres.index[:15],
             color=num_genres.values[:15]) # colorir as barras pela contagem

bar.update_layout(xaxis_title='Gênero', yaxis_title='Número de Aplicativos')
bar.show()
```

```python
bar = px.bar(x = num_genres.index[:15],
             y = num_genres.values[:15],
             title='Top 15 Gêneros na Google Play Store',
             hover_name=num_genres.index[:15],
             color=num_genres.values[:15],
             color_continuous_scale='Plasma') # Adicione esta linha para a escala de cores

bar.update_layout(xaxis_title='Gênero',
                  yaxis_title='Número de Aplicativos',
                  coloraxis_showscale=False) # Adicione esta linha para remover o eixo de cores

bar.show()
```
