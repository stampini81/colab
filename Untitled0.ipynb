```python
import matplotlib.pyplot as plt
import pandas as pd

dados = {'cliente': ["Jonas", "Kim", "Lucas"], 'idade': [15, 39, 68]}
df = pd.DataFrame(dados)

plt.figure(figsize=(6, 4))
plt.bar(df['cliente'], df['idade'])
plt.title('Idade dos Clientes')
plt.xlabel('Cliente')
plt.ylabel('Idade')
plt.show()
```

```python
import pandas as pd

# Caminhos dos arquivos no seu Drive
path1 = '/content/drive/MyDrive/sport-activity.csv'
path2 = '/content/drive/MyDrive/sport-activity2.csv'
path3 = '/content/drive/MyDrive/sport-activity3.csv'

# Carregando os arquivos
df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df3 = pd.read_csv(path3)

# Concatenando os três em um único DataFrame (opcional)
df = pd.concat([df1, df2, df3], ignore_index=True)

# ✅ Excluindo linhas com dados vazios
df = df.dropna()

# ✅ Substituindo valores nulos por 0 (se ainda houver)
df.fillna(0, inplace=True)

# ✅ Verificando e removendo duplicatas
df = df.drop_duplicates()

# ✅ Corrigindo tipos de dados (ex: datas)
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date'])

# ✅ Corrigindo valores específicos (exemplo fictício)
if 'Duration' in df.columns:
    df['Duration'] = df['Duration'].replace(450, 45)

# ✅ Calculando correlação (se tiver colunas numéricas)
correlacoes = df.corr()
print("Correlação entre colunas:")
print(correlacoes)

# ✅ Exibindo o DataFrame final
df.head()
```
