```python
sets_per_theme_name = merged_df['name_theme'].value_counts()
print("Número de conjuntos por Nome do Tema:")
print(sets_per_theme_name.head(10)) # Mostra os 10 principais temas pelo nome
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6)) # Define o tamanho do gráfico
plt.bar(sets_per_theme_name.index[:10], sets_per_theme_name.values[:10]) # Barras para os 10 primeiros
plt.xticks(rotation=45, ha='right') # Rotaciona os rótulos do eixo X para melhor legibilidade
plt.title('Top 10 Temas LEGO por Número de Conjuntos')
plt.xlabel('Nome do Tema')
plt.ylabel('Número de Conjuntos')
plt.tight_layout() # Ajusta o layout para evitar sobreposição de rótulos
plt.show()
```
